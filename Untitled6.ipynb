```python
# ===== 1. Introduction & Setup =====
!pip install scikit-learn pandas numpy matplotlib seaborn -q
import sklearn
sklearn.show_versions()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (mean_absolute_error, mean_squared_error, r2_score,
                             accuracy_score, precision_score, recall_score, f1_score,
                             classification_report, confusion_matrix)

# ===== 2. Load & Explore Dataset =====
# Load dataset from online source
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
           "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
df = pd.read_csv(url, header=None, names=columns)

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nDataset info:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
print("\nMissing values check:")
print(df.isnull().sum())

# ===== 3. Data Preprocessing =====
# Check for zeros in medical measurements (might indicate missing data)
zero_columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
print("\nRows with zeros in critical columns:")
for col in zero_columns:
    zero_count = (df[col] == 0).sum()
    print(f"{col}: {zero_count} zeros ({zero_count/len(df)*100:.2f}%)")

# Replace zeros with NaN and then fill with median
df_clean = df.copy()
for col in zero_columns:
    df_clean[col] = df_clean[col].replace(0, np.nan)
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

print("\nAfter cleaning - no zeros in critical columns:")
print(df_clean[zero_columns].describe().loc[['min', 'max']])

# ===== 4. Visualize Data =====
# Distribution of features
fig, axes = plt.subplots(3, 3, figsize=(12, 10))
axes = axes.flatten()
for i, col in enumerate(columns[:-1]):  # Exclude Outcome
    axes[i].hist(df_clean[col], bins=20, edgecolor='black', alpha=0.7)
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
corr_matrix = df_clean.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# ===== 5. Prepare Data for Models =====
# Split into features (X) and targets (y_regression, y_classification)
X = df_clean.drop(columns=['BloodPressure', 'Outcome'])  # Features for both tasks
y_reg = df_clean['BloodPressure']  # Target for regression
y_clf = df_clean['Outcome']  # Target for classification

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split for both tasks
X_train, X_test, y_reg_train, y_reg_test = train_test_split(
    X_scaled, y_reg, test_size=0.25, random_state=42, shuffle=True
)
X_train_clf, X_test_clf, y_clf_train, y_clf_test = train_test_split(
    X_scaled, y_clf, test_size=0.25, random_state=42, shuffle=True, stratify=y_clf
)

print("Regression - Training set:", X_train.shape, y_reg_train.shape)
print("Regression - Testing set:", X_test.shape, y_reg_test.shape)
print("\nClassification - Training set:", X_train_clf.shape, y_clf_train.shape)
print("Classification - Testing set:", X_test_clf.shape, y_clf_test.shape)

# ===== 6. REGRESSION TASK: Predict Blood Pressure =====
print("\n" + "="*60)
print("REGRESSION TASK: Predicting Blood Pressure")
print("="*60)

# Train Linear Regression model
reg_model = LinearRegression()
reg_model.fit(X_train, y_reg_train)

# Make predictions
y_reg_pred = reg_model.predict(X_test)

# Evaluate regression model
mae = mean_absolute_error(y_reg_test, y_reg_pred)
mse = mean_squared_error(y_reg_test, y_reg_pred)
r2 = r2_score(y_reg_test, y_reg_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.4f}")

# Visualize predictions vs actual
plt.figure(figsize=(8, 6))
plt.scatter(y_reg_test, y_reg_pred, alpha=0.6, edgecolors='k')
plt.plot([y_reg_test.min(), y_reg_test.max()],
         [y_reg_test.min(), y_reg_test.max()], 'r--', lw=2)
plt.xlabel('Actual Blood Pressure')
plt.ylabel('Predicted Blood Pressure')
plt.title('Linear Regression: Actual vs Predicted Blood Pressure')
plt.grid(True, alpha=0.3)
plt.show()

# Feature importance for regression
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': reg_model.coef_
}).sort_values('Coefficient', key=abs, ascending=False)
print("\nFeature Importance for Blood Pressure Prediction:")
print(feature_importance)

# ===== 7. CLASSIFICATION TASK: Predict Diabetes Outcome =====
print("\n" + "="*60)
print("CLASSIFICATION TASK: Predicting Diabetes (Outcome)")
print("="*60)

# 7.1 Logistic Regression
print("\n--- Logistic Regression ---")
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_clf, y_clf_train)
y_log_pred = log_reg.predict(X_test_clf)
y_log_prob = log_reg.predict_proba(X_test_clf)[:, 1]

# Evaluate Logistic Regression
accuracy_log = accuracy_score(y_clf_test, y_log_pred)
precision_log = precision_score(y_clf_test, y_log_pred)
recall_log = recall_score(y_clf_test, y_log_pred)
f1_log = f1_score(y_clf_test, y_log_pred)

print(f"Accuracy: {accuracy_log:.4f}")
print(f"Precision: {precision_log:.4f}")
print(f"Recall: {recall_log:.4f}")
print(f"F1-Score: {f1_log:.4f}")

print("\nClassification Report:")
print(classification_report(y_clf_test, y_log_pred))

# Confusion Matrix for Logistic Regression
cm_log = confusion_matrix(y_clf_test, y_log_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()

# 7.2 K-Nearest Neighbors (KNN)
print("\n--- K-Nearest Neighbors (KNN) ---")
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_clf, y_clf_train)
y_knn_pred = knn.predict(X_test_clf)

# Evaluate KNN
accuracy_knn = accuracy_score(y_clf_test, y_knn_pred)
precision_knn = precision_score(y_clf_test, y_knn_pred)
recall_knn = recall_score(y_clf_test, y_knn_pred)
f1_knn = f1_score(y_clf_test, y_knn_pred)

print(f"Accuracy: {accuracy_knn:.4f}")
print(f"Precision: {precision_knn:.4f}")
print(f"Recall: {recall_knn:.4f}")
print(f"F1-Score: {f1_knn:.4f}")

print("\nClassification Report:")
print(classification_report(y_clf_test, y_knn_pred))

# Confusion Matrix for KNN
cm_knn = confusion_matrix(y_clf_test, y_knn_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Greens',
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - KNN (k=5)')
plt.show()

# ===== 8. Model Comparison =====
print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)

comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNN (k=5)'],
    'Accuracy': [accuracy_log, accuracy_knn],
    'Precision': [precision_log, precision_knn],
    'Recall': [recall_log, recall_knn],
    'F1-Score': [f1_log, f1_knn]
})

print("\nClassification Models Performance:")
print(comparison.to_string(index=False))

# Visualization of model comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, comparison.iloc[0, 1:], width, label='Logistic Regression', color='skyblue')
rects2 = ax.bar(x + width/2, comparison.iloc[1, 1:], width, label='KNN (k=5)', color='lightgreen')

ax.set_ylabel('Score')
ax.set_title('Model Comparison by Evaluation Metrics')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylim(0, 1)
ax.grid(axis='y', alpha=0.3)

# Add value labels
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()

# ===== 9. Additional Analysis: Finding Optimal K for KNN =====
print("\n" + "="*60)
print("FINDING OPTIMAL K FOR KNN")
print("="*60)

# Test different k values
k_values = range(1, 21)
accuracies = []

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_clf, y_clf_train)
    y_pred_temp = knn_temp.predict(X_test_clf)
    accuracies.append(accuracy_score(y_clf_test, y_pred_temp))

# Plot k vs accuracy
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='purple')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN: Finding the Optimal Number of Neighbors')
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.show()

best_k = k_values[np.argmax(accuracies)]
print(f"Optimal k value: {best_k} with accuracy: {max(accuracies):.4f}")

# ===== 10. Summary =====
print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print("1. Regression Task (Blood Pressure Prediction):")
print(f"   - Linear Regression achieved R = {r2:.4f}")
print(f"   - Most important features: {feature_importance.iloc[0, 0]} and {feature_importance.iloc[1, 0]}")

print("\n2. Classification Task (Diabetes Prediction):")
print(f"   - Logistic Regression Accuracy: {accuracy_log:.4f}")
print(f"   - KNN (k=5) Accuracy: {accuracy_knn:.4f}")
print(f"   - Best model: {'Logistic Regression' if accuracy_log > accuracy_knn else 'KNN'}")

print("\n3. Dataset Insights:")
print(f"   - Total samples: {len(df)}")
print(f"   - Diabetes cases: {df['Outcome'].sum()} ({df['Outcome'].mean()*100:.1f}%)")
print(f"   - Average Blood Pressure: {df['BloodPressure'].mean():.1f}")

print("\n Tutorial completed successfully!")
```
